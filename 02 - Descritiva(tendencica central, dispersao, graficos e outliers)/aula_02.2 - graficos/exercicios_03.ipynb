**Exercício 7:**  
Gráfico de Dispersão com Cores - Relação entre Anos de Estudo e Salário por Escolaridade 
(Pernambuco) 

anos_estudo = [8, 12, 16, 6, 10, 18, 11, 14, 9, 15] 

salario = [1500, 2800, 4500, 1200, 2000, 6000, 2200, 3500, 1800, 4000] 

escolaridade = ['Fundamental', 'Médio', 'Superior', 'Fundamental', 'Médio', 'Superior', 'Médio', 
'Superior', 'Fundamental', 'Superior'] 

 
Crie um gráfico de dispersão mostrando a relação entre anos de estudo e salário, com 
cores diferentes para cada nível de escolaridade. Adicione legenda.

---

```python
import numpy as np

x = np.array([8, 12, 16, 6, 10, 18, 11, 14, 9, 15])
y = np.array([1500, 2800, 4500, 1200, 2000, 6000, 2200, 3500, 1800, 4000] )

escolaridade = ['Fundamental', 'Médio', 'Superior', 'Fundamental', 'Médio', 'Superior', 'Médio', 
'Superior', 'Fundamental', 'Superior']

cores = []

for i in escolaridade:
    if i == 'Fundamental':
        i = 'black'
        cores.append(i)
    elif i == 'Médio':
        i = 'green'
        cores.append(i)
    elif i == 'Superior':
        i = 'red'
        cores.append(i)
#print(cores)

colors = np.array(cores)

plt.scatter(x,y,c=colors)
```
