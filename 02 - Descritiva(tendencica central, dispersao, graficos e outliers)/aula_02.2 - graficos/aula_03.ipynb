```python
import matplotlib.pyplot as plt
```

## Grafico em barras

```python
categorias = ['A', 'B', 'C', 'D']
valores = [25, 40, 30, 35]

plt.bar(categorias, valores, color='skyblue')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gráfico de Barras')
plt.show()
```

## Histograma

```python
import numpy as np

dados = np.random.normal(loc=0, scale=1, size=100) 

plt.hist(dados, bins=10, color='lightcoral', edgecolor='black')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma')
plt.show()
```

## Linha

```python
tempo = [1, 2, 3, 4, 5]
temperatura = [20, 22, 25, 23, 26]

plt.plot(tempo, temperatura, marker='o', linestyle='-', color='green')
plt.xlabel('Tempo')
plt.ylabel('Temperatura (°C)')
plt.title('Gráfico de Linha da Temperatura')
plt.grid(True) # Adiciona uma grade ao gráfico
plt.show()
```

## Setor

```python
vendas_por_setor = [45, 30, 15, 10]
rotulos = ['Eletrônicos', 'Vestuário', 'Livros', 'Outros']
cores = ['gold', 'lightskyblue', 'lightcoral', 'lightgreen']
explode = (0.1, 0, 0, 0) # Destaca a primeira fatia

plt.pie(vendas_por_setor, labels=rotulos, colors=cores, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Gráfico de Setor das Vendas')
plt.axis('equal') # Garante que o gráfico seja um círculo
plt.show()
```
